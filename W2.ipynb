```python
import string

def check_password_strength(password):
    errors = []

    if len(password) < 8:
        errors.append("Password must be at least 8 characters long.")

    if not any(char.isupper() for char in password):
        errors.append("Password must contain at least one uppercase letter.")

    if not any(char.islower() for char in password):
        errors.append("Password must contain at least one lowercase letter.")

    special_characters = string.punctuation
    if not any(char in special_characters for char in password):
        errors.append("Password must contain at least one special character (e.g., @, #, $, %).")

    if errors:
        print("Password is weak:")
        for error in errors:
            print("-", error)
    else:
        print("Password is strong")

password = input("Enter your password: ")
check_password_strength(password)
```

```python
def caesar_cipher(message, key, mode):
    result = ""

    for char in message:
        if char.isalpha():  
            shift = key if mode == "encrypt" else -key
            new_char = chr(((ord(char.upper()) - 65 + shift) % 26) + 65)  
            result += new_char if char.isupper() else new_char.lower() 
        else:
            result += char  

    return result

message = input("Enter your message: ")
key = int(input("Enter the shift key (number): "))
mode = input("Enter mode (encrypt/decrypt): ").strip().lower()

if mode not in ["encrypt", "decrypt"]:
    print("Invalid mode! Please enter 'encrypt' or 'decrypt'.")
else:
    output = caesar_cipher(message, key, mode)
    print("Output:", output)
```

```python
import random

def hangman():
    words = ["engine", "brakes", "suspension", "transmission", "battery", "fuel", "radiator", "steering", 
             "clutch", "alternator", "dashboard", "headlights", "taillights", "windshield", "mirrors", 
             "tires", "axle", "chassis", "wheels", "cylinder"]
    
    word = random.choice(words)  
    guessed_word = ["_"] * len(word)  
    max_guesses = 6  
    guessed_letters = set() 
    incorrect_guesses = 0 

    print("Welcome to Hangman - Car Edition!")
    print(" ".join(guessed_word))  

    while incorrect_guesses < max_guesses and "_" in guessed_word:
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.add(guess)

        if guess in word:
            for i, char in enumerate(word):
                if char == guess:
                    guessed_word[i] = guess
            print("Correct!")
        else:
            incorrect_guesses += 1
            print(f"Wrong! You have {max_guesses - incorrect_guesses} guesses left.")

        print(" ".join(guessed_word))  

    if "_" not in guessed_word:
        print(f"Congratulations, you guessed the word '{word}'!")
    else:
        print(f"Game over! The word was '{word}'.")

hangman()
```

```python
import subprocess

def tracert(domain):
    try:
        result = subprocess.run(["tracert", domain], capture_output=True, text=True)  
        print(f"Trace route for {domain}:\n")
        print(result.stdout)
    except FileNotFoundError:
        print("Error: tracert/traceroute command not found. Make sure it's available on your system.")
    except Exception as e:
        print(f"An error occurred: {e}")

domains = ["google.com", "reddit.com", "tiktok.com"]

for domain in domains:
    tracert(domain)
```

```python
import socket
import requests
from bs4 import BeautifulSoup

locations = {
    "Lewisham": "https://www.bbc.co.uk/weather/2644487",
    "King's Cross": "https://www.bbc.co.uk/weather/2648110"
}

weather_data = {}

for location, url in locations.items():
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        temperature_tag = soup.find("span", class_="wr-value--temperature--c")
        if temperature_tag:
            weather_data[location] = int(temperature_tag.text.strip().replace("°", ""))
        else:
            weather_data[location] = None
    else:
        weather_data[location] = None

if None not in weather_data.values():
    if weather_data["Lewisham"] > weather_data["King's Cross"]:
        comparison = "Lewisham is warmer."
    elif weather_data["Lewisham"] < weather_data["King's Cross"]:
        comparison = "King's Cross is warmer."
    else:
        comparison = "Both locations have the same temperature."
    message = (f"Lewisham: {weather_data['Lewisham']}°C, "
           f"King\'s Cross: {weather_data['King\'s Cross']}°C. {comparison}")


else:
    message = "Could not fetch weather data for one or both locations."

client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server_address = ('localhost', 65433)
client_socket.sendto(message.encode(), server_address)
print("Weather data sent!")
client_socket.close()
```

```python
import socket

server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

server_socket.bind(('localhost', 65432))

server_socket.listen(1)
print("TCP Server is listening...")

while True:
    client_socket, client_address = server_socket.accept()
    print(f"Connected to {client_address}")
    
    data = client_socket.recv(1024)
    print(f"Received: {data.decode()}")
    
    client_socket.sendall(b"ACK: " + data)
    
    client_socket.close()
```

Even Parity check

```python
import random

def compute_even_parity(data):
    return sum(data) % 2  

# Function to check even parity
def check_even_parity(data_with_parity):
    data = data_with_parity[:-1]  
    parity = data_with_parity[-1]  
    calculated_parity = sum(data) % 2
    return calculated_parity == parity 

# Original data bits (example)
data = [1, 0, 1, 0, 0, 1, 0, 1]
parity_bit = compute_even_parity(data)
print("Original Data: ", data)
print("Computed Parity Bit (Even):", parity_bit)

transmitted_data = data + [parity_bit]
print("\nTransmitted Data (Data + Parity):", transmitted_data)

error_index = random.randint(0, len(transmitted_data) - 1)
transmitted_data[error_index] = 1 - transmitted_data[error_index] 

print("\nTransmitted Data with Error: ", transmitted_data)
if check_even_parity(transmitted_data):
    print("No error detected.")
else:
    print("Error detected!")
```

Two Dimentinal check

```python
import random

def compute_row_parity(data):
    return sum(data) % 2 

def compute_column_parity(matrix):
    column_parities = []
    num_columns = len(matrix[0])
    for col in range(num_columns):
        column_parity = sum(matrix[row][col] for row in range(len(matrix))) % 2
        column_parities.append(column_parity)
    return column_parities

def check_two_dimensional_parity(matrix, row_parities, column_parities):
    for i in range(len(matrix)):
        if sum(matrix[i]) % 2 != row_parities[i]:
            return False
    computed_column_parities = compute_column_parity(matrix)
    return computed_column_parities == column_parities

matrix = [
    [1, 0, 1],
    [0, 1, 0],
    [1, 1, 1]
]

row_parities = [compute_row_parity(row) for row in matrix]
column_parities = compute_column_parity(matrix)

print("Original Matrix:")
for row in matrix:
    print(row)

print("\nRow Parities:", row_parities)
print("Column Parities:", column_parities)

row_index = random.randint(0, len(matrix) - 1)
col_index = random.randint(0, len(matrix[0]) - 1)
matrix[row_index][col_index] = 1 - matrix[row_index][col_index]  

print("\nMatrix with Error:")
for row in matrix:
    print(row)

if check_two_dimensional_parity(matrix, row_parities, column_parities):
    print("No error detected.")
else:
    print("Error detected!")
```

I experimented with different sets of data, like [1, 0, 1, 0, 1, 1, 0, 0] and [1, 0, 1, 0, 0, 1, 0, 1], to see if the parity check still operates correctly. In both cases, the even parity was computed correctly. I then simulated errors by flipping a random bit and found that the parity check successfully detected the error. So, the check works as expected, even with different data, and can detect errors
